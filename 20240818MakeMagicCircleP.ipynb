# 漸く森博嗣「笑わない数学者」のビリヤードの問題／魔円陣を解いたものの要領を良くする
素数 $p$ に対して,  $A,B,C$ を $0,\ 1,\ \cdots,\ p-1$ で逐次試して, 
+ $C\ne 0$ や $A=B=0$ でない
+ $x=1,...,p-1$ を代入し $x^3+Ax^2+Bx+C$ が $0$ にならない事(既約多項式であること)を確認
+ $(a_0,b_0,c_0)=(0,0,1)$ から始めて, $(a_i,b_i,c_i)$ を漸化式から順に計算し, $i=p^2+p+1$ のとき始めて $a_i=b_i=0$ になることを確認する  

という要領。  
先ずは, 環境の確認と, **Pkg**の導入から

```julia
println(versioninfo())
using Nemo,Combinatorics,Dates,Primes
```

最初の2つの条件を満たす **[A,B,C]** のリストを拵える...と$p$が大きくなったときに効率が悪いので...  
ランダムに **[A,B,C]** を取り上げて... チェックしてから試してみる。

```julia
p=2
n=10
q=p^n
Fpn,α=finite_field(p,n,"α")
Fpnl=[]
for e ∈ Fpn
    push!(Fpnl,e)
end
println(join(Fpnl[1:10],",")," ...")
println(join(Fpnl[end-9:end],","))
function rande()
    l=length(Fpnl)
    Fpnl[rand(1:l)]
end
start = Dates.now()
chk=false
chkA=true
bbl=[]
A,B,C=0,0,0
ABC=[0,0,0]
abclst=[[α,α,α]]
while chk!=true
    A,B,C=rande(),rande(),rande()
    if C!=0 && !(A==0 && B==0)
        chkA=true
        for x ∈ Fpn
            if x^3+A*x^2+B*x+C==0
                chkA=false
                break
            end
        end
        if chkA==true
            chk=true
        end
    end
end
println("next")
if chkA
    ABC=[A,B,C]
    println(ABC)
    abc=[Fpnl[1],Fpnl[1],Fpnl[2]]
    abclst=[abc]
    println(abclst)
    for i=1:q^2+q+1
        nxtabc=[-ABC[1]*abc[1]+abc[2],-ABC[2]*abc[1]+abc[3],-ABC[3]*abc[1]]
#        println(nxtabc)
        push!(abclst,nxtabc)
        if nxtabc[1]==0 && nxtabc[2]==0
            break
        else
            abc=nxtabc
        end
    end
end
chk=true
ABC=[A,B,C]
println(join(ABC,",")," : ",chkA)
prnstrl=[]
for e ∈ abclst[1:5]
    push!(prnstrl,"("*join(e,",")*")")
end
println(join(prnstrl,","))
println("  ...")
prnstrl=[]
for e ∈ abclst[end-4:end]
    push!(prnstrl,"("*join(e,",")*")")
end
println(join(prnstrl,","))
#
cds=[]
for i=1:length(abclst)
    if abclst[i][1]==0
        push!(cds,i-1)
    end
end
bbl=[]
for i=1:length(cds)-1
    push!(bbl,cds[i+1]-cds[i])
end
#println(join(prnstr,",")," : ",length(abclst),", ",p^2+p+1)
println("=> ("*join(cds,",")*")")
println("-> ("*join(bbl,",")*")")
```

```julia
start = Dates.now()
chklst=[]
bbls=[bbl;bbl]
for i=1:length(bbl)-1
    for j=1:length(bbl)
        push!(chklst,sum(bbls[j:j+i-1]))
    end
end
push!(chklst,sum(bbl))
sort!(chklst)
println(join(chklst[1:10],","))
println("   ...")
println(join(chklst[end-9:end],","))
println("p=",p,", n=",n,", q=p^n=",q,", size=",length(bbl),", q^2+q+1=",q^2+q+1,", CheckListの長さ=",length(chklst))
dtime = Dates.now() - start
println("かかった時間: ",dtime.value/1000,"秒だから",dtime.value/60000,"分")
```
