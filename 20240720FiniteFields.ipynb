# 有限体から有限射影平面を構成してみる

```julia
println(versioninfo())
using Nemo,Plots,LaTeXStrings,Latexify,Combinatorics
```

```julia
p,n=3,2
Fpn,α=finite_field(p,n,"α")
println(defining_polynomial(Fpn))
Fpnl=[]
for x ∈ Fpn
    push!(Fpnl,x)
end
println(join(Fpnl,",")," : ",length(Fpnl))
```

### $Fpn^3$ として有限射影平面上の点群を定める。

```julia
Fpnl3=[]
for x ∈ Fpn,y∈Fpn,z∈Fpn
    push!(Fpnl3,[x,y,z])
end
for x ∈ Fpnl3
    print("(",join(x,","),"), ")
end
print(length(Fpnl3))
```

#### $ax+by+cz=0$を満たす点を集めると平面になる？
#### $(0,0,0)$は外してリストにした。

```julia
Fpnl3p=[]
for x ∈ Fpnl3
    a,b,c=x[1],x[2],x[3]
    ln=[]
    for p ∈ Fpnl3
        if a*p[1]+b*p[2]+c*p[3]==0
            if p !=[0,0,0]
                push!(ln,p)
            end
        end
    end
    if x!=[0,0,0] && !(ln ∈ Fpnl3p)
        push!(Fpnl3p,ln)
    end
end
for ln ∈ Fpnl3p
    for x ∈ ln
        print("(",join(x,","),"), ")
    end
    print(" : ",length(ln),"\n")
end
println(length(Fpnl3p))
```

#### 数的にはこんな感じだろうか？
#### 点の集合として平面を逐一洗い出すことはできたようだ。$p^n=q$ として, $q^2+q+1$ と $q^2-1$ を確認しておこう。

```julia
pn=p^n
println(pn^2+pn+1,", ",pn^2-1)
```

#### では直線はどう構成すれば良いだろう。原点を通る直線だから, スカラー倍で良いのか...

```julia
lnbl=[]
for x ∈ Fpnl3
    if x!=[0,0,0]
        lnb=Set()
        for y ∈ Fpn
            if y!=0
                push!(lnb,[y*x[1],y*x[2],y*x[3]])
            end
        end
        if !(lnb ∈ lnbl)
            push!(lnbl,lnb)
        end
    end
end
for l ∈ lnbl
    for p ∈ l
        print("(",join(p,","),"),")
    end
    print("\n")
end
println(length(lnbl),", ",pn^2+pn+1)
```

ちゃんと直線も定義できたようだ。

${F_q}^3$ において
ここで作った直線と平面の集合を $F_q$上の射影平面と呼ぶ。  
$F_q$は位数$q=p^m$の有限体として扱っている。  
${F_q}^3$も有限体の筈だけど, そこは触れてないなぁ... 。
+ $原点を通る直線の数 = q^2 +q +1$
+ $原点を通る平面の数 = q^2 +q +1$
+ $原点を通る 2 平面が含む，原点を通る直線の数 = 1$ 
+ $原点を通る 2 直線を含む，原点を通る平面の数 = 1$

このうちの最初の2つは確認できている。残りの2つを確認しておこう。  
先ずは, 「$原点を通る 2 平面が含む，原点を通る直線の数 = 1$」からだ。

```julia
cnt=0
for comb in combinations(Fpnl3p,2)
    a=Set(comb[1])
    b=Set(comb[2])
    c=intersect(a,b)
    if length(c)==pn-1 && (c ∈ lnbl)
        cnt=cnt+1
    else
        println("Error")
    end
end
pp=pn^2+pn+1
println(cnt,", ",pp*(pp-1)//2)
```

チェックできているようだ。  
次は「$原点を通る 2 直線を含む，原点を通る平面の数 = 1$」だ。

```julia
cnt=0
for comb in combinations(lnbl,2)
    a=Set(comb[1])
    b=Set(comb[2])
    c=union(a,b)
    pcnt=0
    for pln ∈ Fpnl3p
        if issubset(c,pln)
            pcnt=pcnt+1
        end
    end
    if pcnt==1
        cnt=cnt+1
    else
        println("Error")
    end
end
pp=pn^2+pn+1
println(cnt,", ",pp*(pp-1)//2)
```

#### チェック完了

### ここまでで具体的に確認できたこと
#### $p$ を素数として, $p^n$ を位数とする有限体 $F_{p^n}$ を想定する。
#### この $F_{p^n}$ を元に, ${F_{p^n}}^3$ という3次元のベクトル空間を想定し, 射影平面を構成する。
#### つまり, $(x,y,z)\in {F_{p^n}}^3$ を $x,y,z\in F_{p^n}$ として定義して誂えるわけだ。
#### 面倒なので, $q=p^n$ とすれば, 有限体 $F_q$ と, $F_q$ の元を成分とする3次元ベクトル空間 ${F_q}^3$ を設たわけだ。

${F_q}^3$ において, 次の条件を満たす直線と平面の集合ができれば, それを $F_q$上の射影平面と呼ぶ。  
+ $原点を通る直線の数 = q^2 +q +1$
+ $原点を通る平面の数 = q^2 +q +1$
+ $原点を通る 2 平面が含む，原点を通る直線の数 = 1$ 
+ $原点を通る 2 直線を含む，原点を通る平面の数 = 1$

平面については, $a,b,c\in F_q$ に対して $ax+by+cz=0$ を満たす点$(x,y,z)$の集合を平面とする。  
ただし, $a=b=c=0$の場合は省く。
単純に, 組み合わせ的には, $q^3-1$通り想定できるが, 実際には, $q-1$通り重複するので, $$\frac{q^3-1}{q-1}=q^2+q+1$$
通りの平面（点の集合）が想定される。

集合については, $x\in {F_q}^3, x\ne (0,0,0)$に対して, $y=kx, k\in F_q$ を満たす $y\in {F_q}^3$ の集合を直線とする。  
これも沢山ありそうにも思えるかもしれないが, 2直線の共有点はない（原点を含むと設定し直せば, 原点のみ）ので, 平面と同様の数となる。
