### 具体的な $p,\ n$ , $p=2,\ n=3$ の場合の有限(拡大)体を設定し, 生成多項式を確認し, 巡回平面を確認する。
#### Nemo の仕様が変わったのか, Fpn,α=GF(p,n,"α")という書き方はできない。
#### GF(p,n,"α") としても α が計算で使えない。仕方がないので, 有限体の元を配列に取り出し, α に当たるものを計算用に指定している。

```julia
p,n=2,3
Fpn=GF(p,n,"α")
println(defining_polynomial(Fpn))
Fpnl=[]
for x ∈ Fpn 
    print(x,", ")
    push!(Fpnl,x)
end;print("\n")
FpnP,x=[],Fpnl[3]
for i=0:p^2+p+1
    f=x^i
    fpl=[coeff(f,0),coeff(f,1),coeff(f,2)]
    push!(FpnP,fpl)
    println(i+1," : (",join(fpl,","),") : α^",i," = ",f)
end
jps=[[]]
for i=1:p^2+p+1
    if FpnP[i][3]==0
        push!(jps[1],i)
    end
end
m=p^2+p+1
for i=1:p^2+p+1
    njps=jps[i]
    njps=njps.%m
    njps=njps.+1
    push!(jps,sort!(njps))
    println(njps)
end
```

### 同じ事を $p=3,\ n=3$ の場合でも確認してみる。つまり, 生成多項式を確認し, 巡回平面を確認する。

```julia
p,n=3,3
Fpn=GF(p,n,"α")
println(defining_polynomial(Fpn))
Fpnl=[]
for x ∈ Fpn 
    print(x,", ")
    push!(Fpnl,x)
end;print("\n")
FpnP,x=[],Fpnl[4]
for i=1:(p^2+p+1)*(p-1)
    f=x^i
    fpl=[coeff(f,0),coeff(f,1),coeff(f,2)]
    push!(FpnP,fpl)
    println(i+1," : (",join(fpl,","),") : α^",i," = ",f)
end
jps=[[]]
for i=1:(p^2+p+1)*(p-1)
    if FpnP[i][3]==0
        push!(jps[1],i)
    end
end
m=(p^2+p+1)*(p-1)
for i=1:(p^2+p+1)*(p^1)
    njps=jps[i]
    njps=njps.%m
    njps=njps.+1
    push!(jps,sort!(njps))
    println(njps)
end
```

#### $p=3,\ n=3$ の場合は, 巡回平面の巡回が都合3回繰り返される結果となっている。
#### 周期が$13$になっているというわけだ。
