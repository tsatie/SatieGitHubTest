# 素数定理の計算と300桁の素数の数
今更ですが, 素数を暗号に使用して安全性を担保するという話です。ググってみると300桁程の素数であれば, 2つ用意すればその積は600桁ほどになり, これを2つの素因数に分解することは非常に難しいから安全だというのが現状での安全性の担保ということのようです。

そこで, じゃあそんな300桁の素数ってどれくらいあるのだろうか？というのは気になっていたので**Julia**で計算してみようという話。

素数定理で301桁までの素数の大体の個数を計算し, そこから300桁までの素数の大体の数を計算して引けば, 
300桁の素数の大体の数が分かるやんという単純な計算です。

たまにしか**Julia**を使わないので適当です。**big()**と**BigInt()**の使い分け？がよく分かっていませんが結果は出せたようです。

```julia
m=301
x=big(10)^m
y=big(10)^(m-1)
xn=BigInt(ceil(x/log(x)))
yn=BigInt(ceil(y/log(y)))
println(xn-yn)
println(x)
```
