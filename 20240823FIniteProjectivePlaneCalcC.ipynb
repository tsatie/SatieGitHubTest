```julia
#function makeGFwithNemo(p)# 位数 p な Galoir Field を作る。要らんやろこれ。
#    GF(p)
#end
#function makeSetelementList(G)# これも collect で良さげ。
#    Gl=[]
#    for e ∈ G
#        push!(Gl,e)
#    end
#    Gl
#end
all_perm(x, n) = vec(map(collect, Iterators.product(ntuple(_ -> x, n)...)))# 配列 x の要素を n 個並べた「重複順列」の配列を作る
#
p,n=3,2# 具体例の為の p,n の設定
p,n=3,4# 具体例の為の p,n の設定
p,n=3,5# 具体例の為の p,n の設定
#GFp=GF(p)#makeGFwithNemo(p)$ GF(p)を拵えて, GFp とする # Nemo.jl を利用するなら
GFp=Vector(0:p-1)# 整数で処理するならこれで
#GFpl=makeGFelementList(GFp)
#GFpl=collect(GFp)# Nemo を利用した場合は, GFp は配列ではないので, 配列にしたものを GFpl としておく
#println(GFpl)# Nemo を利用した場合に
println(join(GFp,","))# GFp の元を配列な　GFpl で表示する。
#GFpx,x = GFp["x"]
IrreduciblePolynomialList=[]# 既約多項式の係数配列を入れる配列
coeflst=all_perm(GFp, n)# 重複順列を利用して, GFpを係数とする n 次多項式の n-1 次以下の係数候補を配列で拵える。
#                         n次の係数は 1 だからそこは拵えずに処理を進めている。
for j=1:length(coeflst)
    prnstrl=[]
    chk=true
    for x=0:p-1
        prnstr=string(x)*"^"*string(n)
        val=x^n
        for i=1:n
            val=(val+coeflst[j][i]*x^(n-i))%p
            val=val
            if coeflst[j][i]!=0
                prnstr=prnstr*"+"*string(coeflst[j][i])*"×"*string(x)*"^"*string(n-i)
            end
        end
        if val!=0
            push!(prnstrl,"f("*string(x)*")="*prnstr*"≡"*string(val))
        else
            chk=false
            break
        end
    end
    if chk
        push!(IrreduciblePolynomialList,coeflst[j])
        println(coeflst[j])
        println(join(prnstrl,", "))
    end
end
#println(join(IrreduciblePolynomialList,", "))
prnstrl=[]
for cl ∈ IrreduciblePolynomialList
    push!(prnstrl,"("*join(cl,",")*")")
end
println(join(prnstrl,", "))
```

#### 上の既約多項式候補をリストアップする部分をルーチン化しておこう。

```julia
function IrreduciblePolynomialListMaker(p,n)# GF(p)をGF(p^n)に拡大する際に利用する既約多項式をリストアップする
    GFp=Vector(0:p-1)# 整数で処理するための, GF(p) は整数の配列で設定
    coeflst=all_perm(GFp, n)# 重複順列を利用して, GFpを係数とする n 次多項式の n-1 次以下の係数候補を配列で拵える。
#                             n次の係数は 1 だからそこは拵えずに処理を進めている。
    ansl=[]
    for j=1:length(coeflst)
#        prnstrl=[]
        chk=true
        for x=0:p-1
#            prnstr=string(x)*"^"*string(n)
            val=x^n
            for i=1:n
                val=(val+coeflst[j][i]*x^(n-i))%p
#                val=val
#                if coeflst[j][i]!=0
#                    prnstr=prnstr*"+"*string(coeflst[j][i])*"×"*string(x)*"^"*string(n-i)
#                end
            end
            if val!=0
#                push!(prnstrl,"f("*string(x)*")="*prnstr*"≡"*string(val))
            else
                chk=false
                break
            end
        end
        if chk
            push!(ansl,coeflst[j])
#            println(coeflst[j])
#            println(join(prnstrl,", "))
        end
    end
    ansl
end
IPlst=IrreduciblePolynomialListMaker(p,n)
println(IPlst)
```

#### 次にこれらの既約多項式の中から, 生成多項式として利用できるもの, つまり, 根の累乗が GF(p^n) を構成するものをリストアップする。

```julia
function PrimitivePolynomialListMaker(p,n)
    IPlst=IrreduciblePolynomialListMaker(p,n)
    ansl=[]
    for cl ∈ IPlst
#    println(" ---------- ")
        Alp=zeros(Int64,n, n)
        for i=2:n
            Alp[i-1,i]=1
        end
        for i=1:n
            Alp[i,1]=(p-cl[i])%p
        end
        be=Vector(1:n).*0
        be[n]=1
        e=be
#    println(e)
    chk=true
        for i=1:p^n-2
            e=(Alp*e).%p
#        println(e)
            if e==be
                chk=false
                break
            end
        end
        if chk
            push!(ansl,cl)
#            println(cl)
#        display(Alp)
#        println("*** OK ***")
        end
    end
    ansl
end
PPlst=PrimitivePolynomialListMaker(p,n)
println(PPlst)
```

こうして得られた, 原始元(生成元？)を根とする既約多項式(原始多項式？生成多項式？)から一つ選んで, 拡大体を構成すれば良い。  
$GF(p^n)$の構成元(配列表示)を確認してみよう。

```julia
function makeGFpnElementList(p,cl)# GF(p) を 係数配列 cl な原始多項式(但し, 最高次な n 次の係数 1 は配列外) で作成した 
#                                   拡大体 GF(p^n) の元を, 配列の配列で列挙する。
#                                   n は原始多項式 cl の係数配列の「大きさ/長さ」になっているので指定は不要な点に注意
    n=length(cl)
    Alp=zeros(Int64,n, n)
    for i=2:n
        Alp[i-1,i]=1
    end
    for i=1:n
        Alp[i,1]=(p-cl[i])%p
    end
    be=Vector(1:n).*0
    ansl=[Vector(1:n).*0]
    be[n]=1
    e=be
    push!(ansl,e)
#    println(e)
    chk=true
    for i=1:p^n-2
        e=(Alp*e).%p
#        println(e)
        if e==be
            chk=false
            println(" *** Break ***")
            break
        else
            push!(ansl,e)
        end
    end
    if !chk
        ansl=[]
    end
    ansl
end
elst=makeGFpnElementList(p,PPlst[2])
println(elst)
println(length(elst),",",length(unique(elst)))
```

#### あとの計算の為に, 拡大で利用した原始多項式の根　$\alpha$ の累乗な $\alpha^i$ を返してくれる関数を拵えておこう。

```julia
function GFpnElement(p,cl,i)# 原始多項式の原始根 α の累乗 α^i を配列で返してくれる。
    n=length(cl)
    be=Vector(1:n).*0
    be[n]=1
    if i==0
        be
    else
        Alp=zeros(Int64,n, n)
        for i=2:n
            Alp[i-1,i]=1
        end
        for i=1:n
            Alp[i,1]=(p-cl[i])%p
        end
        e=be
        for i=1:i
            e=(Alp*e).%p
        end
        e
    end
end
prnlst=[]
for i=0:10
push!(prnlst,GFpnElement(p,PPlst[2],i))
end
println(join(prnlst,","))
```

#### kα^r×lα^s を処理して配列で返すものを拵えよう
その為には, $GF(p)$ を $GF(p^n)$ に拡大するときに, 既約多項式は何を使っているかも引数に入れておく事にしよう。

```julia
ppnum=2# 原始多項式 PrimitivePolynomial の番号
function gfpCalcMulB(p,n,k,r,l,s,cl)# GF(p^n) で, kα^r × lα^s を計算して配列で返す
# 結果は当然, k×l×α^{r+s} となるわけだ。これを a α^b としよう
    a=(k*l)%(p)
    b=(r+s)# ここを最初勘違いして p にしてもうてた
    c=GFpnElement(p,cl,b)
    (a*c).%p
end
println(gfpCalcMulB(p,n,2,3,4,2,PPlst[ppnum]))
println(gfpCalcMulB(p,n,1,0,1,0,PPlst[ppnum]))
println(gfpCalcMulB(p,n,1,1,1,0,PPlst[ppnum]))
println(gfpCalcMulB(p,n,1,1,1,1,PPlst[ppnum]))
println(gfpCalcMulB(p,n,1,1,1,2,PPlst[ppnum]))
println(gfpCalcMulB(p,n,1,1,1,3,PPlst[ppnum]))
```

#### p=3,n=4 で [2,1,1,2] の2乗の場合はどう計算すれば良いだろう。
これも先ずは汎用に, [a,b,c,...,d]と[e,f,g,...,h]の場合の計算を仕込むべきか。

```julia
function gfpCalcMulA(p,n,abcd,efgh,irrpolc)# GF(p^n) で, [a,b,c,...,d] × [e,f,g,...,h] を計算して配列で返す
#                                            つまり, (aα^n-1+bα^n-2+cα^n-3+...+d)(eα^n-1+fα^n-2+gα^n-3+...+h) を計算して係数配列で返す
    ansl=Vector(1:n).*0
    for i=1:n,j=1:n
        ansl=(ansl+gfpCalcMulB(p,n,abcd[i],n-i,efgh[j],n-j,irrpolc)).%p
    end
    ansl
end
```

```julia
function IrreduciblePolynomial3checker(p,n,GFpn,cll,irrpolc)# 3次の既約多項式かどうかチェック... あまま
    ans=true
    zerol=Vector(0:p-1).*0
    for xl ∈ GFpn
        vall=gfpCalcMulN(p,n,xl,3,irrpolc)
        for i=1:3
            vall=vall+gfpCalcMulA(p,n,cll[i],gfpCalcMulN(p,n,xl,3-i,irrpolc),irrpolc)
        end
        if vall==zerol
            ans=false
            break
        end
    end
    ans
end
# ------------------------------------------------------------
p,n=3,4# 具体例の為の p,n の設定
println("p=",p,", n=",n," で有限体 GF(p) から, 原始多項式を一つ選んで, GF(p^n)を拵える")
GFp=Vector(0:p-1)# 整数で処理するならこれで
#println(join(GFp,","))# GFp の元を配列な　GFpl で表示する。
PPlst=PrimitivePolynomialListMaker(p,n)# GF(p)の n 次の原始多項式の一覧を
println(PPlst)
ppnum=1
GFpn=makeGFpnElementList(p,PPlst[ppnum])
```

#### こうして得られた $GF(p^n)$の元（配列）を係数とする 3次多項式（但し, 3次の係数は 1 固定）から行列を...

```julia
function gfpMake3Matrix(p,n,cll)# GF(p^n) n次配列係数の3次の多項式 cll から, ωの累乗用の 3×3行列を用意する
    e=Vector(1:n).*0# n次の 0配列を用意する
    pv=(Vector(1:n).*0).+p# n次の要素が全て p の配列を用意する
    Alps=[e,e,e]# 3×3行列の「列」を用意する。
    Alpp=[Alps Alps Alps]# それを3つ並べて 3×3の 0行列を用意する。
    ee=copy(e)# 最初に用意した 0配列をコピーする
    ee[n]=1# 第 n 成分のみを 1 にする。
    for i=2:3
        Alpp[i-1,i]=ee
    end
    for i=1:3
        Alpp[i,1]=(pv-cll[i]).%p
    end
    Alpp
end
Alpp=gfpMake3Matrix(p,n,cll)
display(Alpp)
```

```julia
function PrimitivePolynomial3Picker(p,n,GFpn,irrpolc)# 3次の原始多項式をピックアップ。乱数頼りで
    GFpnlen=length(GFpn)
    chkA=false
    cll=[]
    while chkA==false
        cll=[GFpn[rand(1:GFpnlen)],GFpn[rand(1:GFpnlen)],GFpn[rand(1:GFpnlen)]]
        if IrreduciblePolynomial3checker(p,n,GFpn,cll,PPlst[ppnum])
#            display(cll)
            println(cll)
            Alpp=gfpMake3Matrix(p,n,cll)
            display(Alpp)
#        println("乱数で係数を決めてOKが出た既約多項式 cll = ",cll)
            es=Vector(1:n).*0
            eo=copy(es)
            eo[n]=1
            be=[es,es,eo]
            e=be
            println("> ",e)
            for g=1:(p^n)^3+2
                e=gfpCalcMatVec(p,n,Alpp,e,PPlst[ppnum])
                if (g-1)%((p^n)^2)==0
                    println(g-1,"/",(p^n)^3-1," : ",e)
                end
                if (e[1]==es) && (e[2]==es) && (e[3]==eo)
                    println(g," : ",e," ",(p^n)^3-1)
                    if g>=(p^n)^3-1
                        chkA=true
#                        println("*** OK ***")
                    else
                        println("*** ",g," *** Failed ***")            
                    end
                    break
                end
            end
            println(" *** fin. ***")
#            println((p^n)^3)
        end
    end
    cll
end
println("p=",p,", n=",n," で有限体 GF(p) から, 原始多項式を一つ選んで, GF(p^n)を拵える")
ansl=PrimitivePolynomial3Picker(p,n,GFpn,PPlst[ppnum])
println(ansl)
```
